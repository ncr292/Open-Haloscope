Circuit simulation of the Open Haloscope hardware

```python
import numpy as np
import matplotlib.pyplot as plt

# lcapy
from lcapy import Circuit, j, omega, s
```

```python
cct = Circuit( """
               Vi 1 0_1 step; down
               R 1 2; right=1.5
               C 2 0; down
               W 0_1 0; right
               W 0 0_2; right=0.5
               P1 2_2 0_2; down
               W 2 2_2;right=0.5
               """ )
```

```python
cct.draw()
```

```python
H = (cct.C.V(s) / cct.Vi.V(s)).simplify()
```

```python
H1 = H.subs('C',1e-6).subs('R',1e3)
H1(j * omega)
```

```python
w = np.logspace(1, 6, 500)

ax = H1(j * omega).dB.plot(w, log_frequency=True)
ax = H1(j * omega).phase_degrees.plot(w, log_frequency=True)
```
